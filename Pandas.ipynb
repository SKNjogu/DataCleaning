# Multiple variables are stored in one column

```python
ebola = pd.read_csv("../data/country_timeseries.csv")
ebola.head()
```

```python
ebola_tidy = ebola.melt(id_vars= ["Date", "Day"],
                       var_name= "cd_country",
                       value_name = "count")
ebola_tidy
```

```python
#Splitting column
#'hello_world'.split('_')
ebola_split = ebola_tidy['cd_country'].str.split('_',expand= True)
```

```python
ebola_tidy[['status','country']] = ebola_tidy['cd_country'].str.split('_',expand= True)
```

```python
#Another way 
ebola_tidy[['status','country']] = ebola_split
```

# Variables are stored in both rows and columns

```python
weather = pd.read_csv("../data/weather.csv")
weather.head()
```

```python
#
weather_tidy= weather.melt(id_vars=["id","year","month", "element"],
                          var_name = "day",
                          value_name = "temp")
```

```python
weather_tidy.head()
```

```python
#Convert element column to individual columns
#Drops missing values here
(weather_tidy.pivot_table(index=["id","year","month","day"],
                        columns="element",
                        values="temp")
 .reset_index()
)
```
